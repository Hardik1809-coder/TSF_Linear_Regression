# Prediction using Supervised ML

### Importing the required libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

### Reading the csv file

```python
data=pd.read_csv("student_scores - student_scores.csv")
```

```python
data.head()
```

### Visualization of the dataset

```python
fontdict={"fontname":"Arial","fontsize":15}
plt.figure(figsize=(6,4),dpi=110)
plt.plot(data["Scores"],"r*-",label="Scores",markersize=10)
plt.title("Variation of Scores of different students",fontdict=fontdict)
plt.xlabel("Roll Number",fontdict=fontdict)
plt.ylabel("Scores",fontdict=fontdict)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=110)
plt.scatter(data.Hours,data.Scores)
plt.title("Variation of Scores with Hours",fontdict=fontdict)
plt.xlabel("Hours",fontdict=fontdict)
plt.ylabel("Scores",fontdict=fontdict)
plt.show()
```

### Creating the object of model class

```python
model=LinearRegression()
```

### Splitting the dataset into attributes and labels

```python
X = data.iloc[:, :-1].values 
y=data.iloc[:,1:].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_train,y_train)
```

### Training the model

```python
model.fit(X_train,y_train)
```

### Visualization of the linear regression line after training

```python
plt.figure(figsize=(6,4),dpi=110)
line = model.coef_*X+model.intercept_
plt.scatter(X, y)
plt.plot(X, line,"g--")
plt.title("Linear Regressor Line",fontdict=fontdict)
plt.xlabel("Hours",fontdict=fontdict)
plt.ylabel("Scores",fontdict=fontdict)
plt.show()
```

```python
print("The testing data:\n",X_test)
```

### Making Predictions

```python
y_pred=model.predict(X_test)
```

```python
print("The predicted values:\n",y_pred)
```

```python
df=pd.DataFrame({"Actual":y_test.reshape(5),"Predicted":y_pred.reshape(5)})
print("Comparing the actual and predicted value")
df
```

### Predicting with random values

```python
hrs=[[9.25]]
```

```python
predict=model.predict(hrs)
```

```python
print("If a Student studies for {} hours then the predicted score will be {} ".format(hrs[0][0],predict[0,0]))
```

### Calculating the accuracy of the model

```python
mae=metrics.mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error:",mae)
```

```python
mse=metrics.mean_squared_error(y_test,y_pred)
print("Mean Squared Error:",mse)
```
